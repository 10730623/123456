```python
import numpy as np
import matplotlib.pyplot as plt
width=0.001
x=np.arange(0,1,width) # interval from 
y=x**2
# plot graph
plt.plot(x,y)
plt.xlim(0,1)  
plt.ylim(0,1) 
# set the xlim to left, right
plt.show()
```
